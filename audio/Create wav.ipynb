```python
import struct
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from scipy import signal
```

```python
sampling_frequency =44100
# sampling_frequency = 500


seconds = 20
# seconds = 8.192
samples = sampling_frequency*seconds
samples_array = np.arange(samples)
signal_amplification_percentage = 100

##calculating time_stamp
time_step = 1/sampling_frequency
time_stamp = (time_step)*samples_array

##sin wave
signal_frequency = 1000
sin_wave = np.sin(2*np.pi*signal_frequency*(samples_array/sampling_frequency))
```

```python
## To plot the waveform

test_signal = sin_wave

fig = plt.figure(figsize=(16,4))
plt.plot(time_stamp,test_signal)
plt.xlabel('Time (s)')
plt.ylabel('Signal (arbitrary units)')


plt.show()
```

```python
##Verifyig signal with FFT

from scipy.fftpack import fft, fftshift, fftfreq, ifft
import numpy as np
from matplotlib.mlab import psd, csd

fig = plt.figure(figsize=(10,8))

nfft = len(test_signal)

signal_array = np.array(test_signal) #arbitrary units
signal_array = np.nan_to_num(signal_array)


#getting the power spectral density -- converts to real numbers (usually in log scale)
[signal_psd, frequency_spectrum] = psd(signal_array, NFFT = nfft, Fs=sampling_frequency)


#plotting:
plt.plot(frequency_spectrum,signal_psd)
plt.title('FFT of Audio Signal')
plt.xlabel("Frequency(Hz)")
plt.ylabel("Amplitude")
plt.yscale('log')
plt.xscale('log')

# plt.xlim([0,30])

plt.grid()

plt.show()
```
